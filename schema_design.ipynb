```python
import pandas as pd

df = pd.read_csv("./Datasets/cleaned_healthcare_claims.csv")
df.head()
```

### Members Table

```python
members = df[[
    "PatientID",
    "PatientAge",
    "PatientGender",
    "PatientIncome",
    "PatientMaritalStatus",
    "PatientEmploymentStatus"
]].drop_duplicates()

members = members.rename(columns={
    "PatientID": "MemberID",
    "PatientAge": "Age",
    "PatientGender": "Gender",
    "PatientIncome": "Income",
    "PatientMaritalStatus": "MaritalStatus",
    "PatientEmploymentStatus": "EmploymentStatus"
})
```

```python
members.sample(5)
```

### Providers Table

```python
providers = df[[
    "ProviderID",
    "ProviderSpecialty",
    "ProviderLocation"
]].drop_duplicates()

providers = providers.rename(columns={
    "ProviderSpecialty": "Specialty",
    "ProviderLocation": "Location"
})
```

```python
providers.sample(5)
```

### Claims Table

```python
claims = df[[
    "ClaimID",
    "PatientID",
    "ProviderID",
    "ClaimAmount",
    "ClaimDate",
    "DiagnosisCode",
    "ClaimStatus",
    "ClaimType",
    "ClaimSubmissionMethod"
]].copy()

claims = claims.rename(columns={
    "PatientID": "MemberID",
    "ClaimSubmissionMethod": "SubmissionMethod"
})
```

```python
claims.sample()
```

### Date conversion

```python
claims["ClaimDate"] = pd.to_datetime(claims["ClaimDate"])
```

### Services Table

```python
services = df[[
    "ClaimID",
    "ProcedureCode"
]].copy()

services["ServiceID"] = range(1, len(services) + 1)
```
